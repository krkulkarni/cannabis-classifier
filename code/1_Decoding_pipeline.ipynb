# 1. Data preparation

```python
# Create the labels and training arrays
# Note that each element is duplicated to account for the splitting of the parcellation into 2 runs later

splits_df = pd.read_csv(f'{data_dir}/mj_splits_training_testing.csv')
labels    = np.repeat(np.array(splits_df['label']), 2)
training  = np.repeat(np.array(splits_df['training']), 2)
names     = np.repeat(np.array(splits_df['subject']), 2)
```

### Functional connectivity

```python
outfile = f'{feat_mat_dir}/standard_fcs_labels_training'
if os.path.isfile(outfile + '.npz'):
    standard_fcs = np.load(outfile + '.npz')
else:
    raw = []
    for sub_name in splits_df['subject']:
        full_ts = pd.read_csv(f'{parcellation_dir}/{sub_name}_stanford_rois.csv', sep='\t', header=None).values.T
        raw.append(full_ts)

    standard_fcs = np.zeros((vector_size, len(raw)*2))
    for i, (tseries, sub_name) in enumerate(zip(raw, splits_df['subject'])):
        clear_output(wait=True)
        print(f'{np.round((i+1)/len(raw)*100, 3)}% complete')
        half = int(tseries.shape[1]/2)
        standard_fcs[:,2*i]   = symm_mat_to_ut_vec(1-pairwise_distances(tseries[:, 0:half], metric='correlation')) 
        standard_fcs[:,2*i+1] = symm_mat_to_ut_vec(1-pairwise_distances(tseries[:, half:], metric='correlation'))
    np.savez(outfile, standard_fcs=standard_fcs.T, labels=labels, training=training)
```

### Functional activity

```python
files = ['mj_cue_betas_labels_training','ctl_cue_betas_labels_training',
         'mj_ctl_contrast_labels_training','mj_cue_betas_mds_labels_training',
         'ctl_cue_betas_mds_labels_training','mj_ctl_contrast_mds_labels_training']
if np.sum([os.path.isfile(f'{feat_mat_dir}/{file}.npz') for file in files]) == len(files):
    print('All files exist.')
else:
    mj_cue_betas = np.zeros((num_subs*2, num_parcels))
    ctl_cue_betas = np.zeros((num_subs*2, num_parcels))
    mj_ctl_contrast = np.zeros((num_subs*2, num_parcels))

    mj_cue_betas_mds = np.zeros((num_subs*2, vector_size))
    ctl_cue_betas_mds = np.zeros((num_subs*2, vector_size))
    mj_ctl_contrast_mds = np.zeros((num_subs*2, vector_size))

    for s, sub_name in enumerate(splits_df['subject']):

        clear_output(wait=True)
        print(f'{np.round((s+1)/ num_subs*100, 3)}% complete')

        beta_means = pd.read_csv(f'{betas_dir}/{sub_name}_beta_parcellation.tsv',sep="\t").T

        mj_cue_betas[2*s,:]       = beta_means.loc[:,0].values 
        mj_cue_betas[2*s+1,:]     = beta_means.loc[:,5].values
        mj_cue_betas_mds[2*s,:]   = make_mean_distance_ut_vec(mj_cue_betas[2*s,:])
        mj_cue_betas_mds[2*s+1,:] = make_mean_distance_ut_vec(mj_cue_betas[2*s+1,:])

        ctl_cue_betas[2*s,:]       = beta_means.loc[:,2].values 
        ctl_cue_betas[2*s+1,:]     = beta_means.loc[:,7].values
        ctl_cue_betas_mds[2*s,:]   = make_mean_distance_ut_vec(ctl_cue_betas[2*s,:])
        ctl_cue_betas_mds[2*s+1,:] = make_mean_distance_ut_vec(ctl_cue_betas[2*s+1,:])

        mj_ctl_contrast_mds[2*s,:]   = make_mean_distance_ut_vec(mj_ctl_contrast[2*s,:])
        mj_ctl_contrast_mds[2*s+1,:] = make_mean_distance_ut_vec(mj_ctl_contrast[2*s+1,:])        
        mj_ctl_contrast[2*s,:]       = mj_cue_betas[2*s,:]   - ctl_cue_betas[2*s,:]
        mj_ctl_contrast[2*s+1,:]     = mj_cue_betas[2*s+1,:] - ctl_cue_betas[2*s+1,:]


    ## save
    outfile = f'{feat_mat_dir}/mj_cue_betas_labels_training'
    np.savez(outfile, mj_cue_betas=mj_cue_betas, labels=labels, training=training)

    outfile = f'{feat_mat_dir}/mj_ctl_contrast_labels_training'
    np.savez(outfile, mj_ctl_contrast=mj_ctl_contrast, labels=labels, training=training)

    outfile = f'{feat_mat_dir}/mj_cue_betas_mds_labels_training'
    np.savez(outfile, mj_cue_betas_mds=mj_cue_betas_mds, labels=labels, training=training)

    outfile = f'{feat_mat_dir}/mj_ctl_contrast_mds_labels_training'
    np.savez(outfile, mj_ctl_contrast_mds=mj_ctl_contrast_mds, labels=labels, training=training)## save
```

# 2. Training

### 2a. Hyperparameter search for connectivity-based models
    Logistic regression + support vector machine grid search, over a range of hyperparameters
    
    (The clf grid and param grid is in the utils file)

```python
# get the data, labels
infile = np.load(f'{feat_mat_dir}/standard_fcs_labels_training.npz')
labels = infile['labels']
inputs = infile['standard_fcs']
training = infile['training']
X_train, X_test = inputs[training==1], inputs[training==0]
y_train, y_test = labels[training==1], labels[training==0]

# run & output grid search
cl = ClassifierGridSearch(clf_grid, param_grid) # initialize grid search instance
cl.grid_search(X_train, y_train, cv=10) # search over training data
sgd_results = pd.DataFrame(cl.grid_searches['SGDClassifier'].cv_results_)
out_name = f'{analysis_dir}/standard_fcs_sgd_hyperparamter_training.xlsx'
sgd_results.to_excel(out_name)
```

### 2b. Hyperparameter search for activation-based models


```python
distance_metrics = ['mj_ctl_contrast',]
for distance_metric in distance_metrics: 
    
    infile = np.load(f'{feat_mat_dir}/'+distance_metric+'_labels_training.npz')
    
    labels = infile['labels']
    inputs = infile[distance_metric]
    training = infile['training']
    X_train, X_test = inputs[training==1], inputs[training==0]
    y_train, y_test = labels[training==1], labels[training==0]

    cl = ClassifierGridSearch(clf_grid, param_grid)
    cl.grid_search(X_train, y_train, cv=10)

    sgd_results = pd.DataFrame(cl.grid_searches['SGDClassifier'].cv_results_)
    out_name = f'{gridsrch_dir}/' + distance_metric + '_sgd_hyperparamter_training.xlsx'
    sgd_results.to_excel(out_name)
```

### 2c. Model comparison: activation models compared to permutation accuracy distribution 

```python
# compare the best performer to a permuted distribution - trained same way, in training set only

best_ix = np.where(sgd_results['rank_test_score'] == 1)[0][0] # just take first one
cv_acc = sgd_results.iloc[best_ix]['mean_test_score']
params = sgd_results.iloc[best_ix]['params']

best_clf = SGDClassifier(random_state=random_state, max_iter=max_iter,
                            penalty = params['penalty'],
                            alpha = params['alpha'],
                            loss = params['loss'])

perm_acc_dist = clf_permutation_cv(best_clf, X_train, y_train, num_folds=10, num_models=1000)
perm_accs_df = pd.DataFrame(perm_acc_dist)
# perm_accs_df.to_csv(gridsrch_dir + '/permutations/Permuted_by_sub_'+ distance_metric +'clf_accuracies.csv')
```

### 3b. Re-train models

```python
infile   = np.load(f'{feat_mat_dir}/standard_fcs_labels_training.npz')
labels   = infile['labels']
inputs   = infile['standard_fcs']
training = infile['training']
X_train, X_test = inputs[training==1], inputs[training==0]
y_train, y_test = labels[training==1], labels[training==0]
```

```python
max_iter = 100000
random_state = 1
alpha = 0.0001
class_weight = None

classifiers = {
    'L1_logistic': SGDClassifier(loss='log', penalty='l1',
                                random_state=random_state, 
                                class_weight=class_weight,
                                max_iter=max_iter),
    'L2_logistic': SGDClassifier(loss='log', penalty='l2',
                                random_state=random_state, 
                                class_weight=class_weight,
                                max_iter=max_iter, 
                                alpha=alpha),
    'L1_LinearSVC': SGDClassifier(loss='hinge', penalty='l1',
                                random_state=random_state,
                                class_weight=class_weight,
                                max_iter=max_iter),
    'L2_LinearSVC': SGDClassifier(loss='hinge', penalty='l2',
                                random_state=random_state, 
                                class_weight=class_weight,
                                max_iter=max_iter, 
                                alpha=alpha)
    }
```

```python
cl = ClassifierPipeline(classifiers)
cl.fit(X_train, y_train, numfolds=10)
```

### 3c. Out-of-sample testing

```python
print(f'\t\tAccuracy\tAUC\tPrecision\tRecall')

weights   = np.zeros((len(cl.classifier_dict), 4005))
preds     = np.zeros((len(cl.classifier_dict), len(X_test)))
test_acc  = np.zeros(len(cl.classifier_dict))
aucs      = np.zeros(len(cl.classifier_dict))
precision = np.zeros(len(cl.classifier_dict))
recall    = np.zeros(len(cl.classifier_dict))

output_dict = {}
for c, (clf_name, classifier) in enumerate(classifiers.items()):
    
    weights[c,:]   = classifier.coef_
    output_dict[clf_name] = classifier.coef_
    preds[c,:]     = classifier.predict(X_test)
    test_acc[c]    = classifier.score(X_test, y_test)*100    
    fpr, tpr, thresholds = roc_curve(y_test, preds[c,:], pos_label=1)
    aucs[c] = auc(fpr, tpr)
    precision[c], recall[c], _, _ = precision_recall_fscore_support(y_test, preds[c,:], average='weighted')
    
    print(f'{clf_name}:\t{np.round(test_acc[c], 2)}%\t\t{np.round(aucs[c], 3)}\t{np.round(precision[c], 3)}\t\t{np.round(recall[c], 3)}')
    
# pickle_file(output_dict, analysis_dir + '/Model_weights.pkl')
```

# 4. Model comparison
    The 4 final models are highly similar in terms of the model weights and predictions

### 4a. Weight similarity

```python
clf_names = [clf_name for clf_name,_ in classifiers.items()]
combos = list(combinations([0,1,2,3], 2))
for combo in combos:
    coef, _ = scipy.stats.kendalltau(weights[combo[0]], weights[combo[1]])
    print(clf_names[combo[0]],clf_names[combo[1]], np.round(coef,3))
```

### 4b. Prediction similarity

```python
# get predictions into matrix
pred_matrix = np.zeros((len(X_test), len(cl.classifier_dict)+2))
for c, (cname, classifier) in enumerate(cl.classifier_dict.items()):
    pred_matrix[:,c] = classifier.predict(X_test)
pred_matrix[:,c+1] = np.mean(pred_matrix[:,1:-2],1) # ensemble average 
pred_matrix[:,c+2] = (pred_matrix[:,c+1] > .49).astype(int) # ensemble prediction (>=.50)
pred_matrix = np.vstack([y_test, pred_matrix.T]).T # add true labels
sorted_ind = np.argsort(pred_matrix[:,0]) # sort by label

# plot
plt.figure(figsize=(5*len(pred_matrix[1,:]), 5))
y_tick_labels = ['True Labels'] + list(cl.classifier_dict.keys()) + ['Ensemble Average'] + ['Ensemble Prediction']
# sns.heatmap(pred_matrix.T[:,sorted_ind], yticklabels=y_tick_labels, xticklabels=np.array(names)[training==0][sorted_ind],
sns.heatmap(pred_matrix.T[:,sorted_ind], yticklabels=y_tick_labels,
            cmap='Blues', linewidths=.25, linecolor='black', cbar=False)
plt.yticks(rotation=60, fontsize = 8)
plt.show()
```

```python
combos = list(combinations([1,2,3,4], 2))
for combo in combos:
    j = jaccard_score(pred_matrix.T[combo[0],sorted_ind], pred_matrix.T[combo[1],sorted_ind])
    print(y_tick_labels[combo[0]], y_tick_labels[combo[1]], np.round(j,3))
```
